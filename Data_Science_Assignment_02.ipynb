# Which movie made the highest profit? Who were its producer and director? Identify the actors in that film.

```python
def extract_actors(x):
    actors = []
    try:
        data = ast.literal_eval(x)
        for i in data:
            actors.append(i.get('name'))
    except:
        pass
    return actors

df['actors'] = df['cast'].apply(extract_actors)
```

```python
def extract_director(x):
    try:
        data = ast.literal_eval(x)
        for i in data:
            if i.get('job') == 'Director':
                return i.get('name')
    except:
        return None

df['director'] = df['crew'].apply(extract_director)
```

```python
def extract_producers(x):
    producers = []
    try:
        data = ast.literal_eval(x)
        for i in data:
            if i.get('job') == 'Producer':
                producers.append(i.get('name'))
    except:
        pass
    return producers

df['producers'] = df['crew'].apply(extract_producers)
```

```python
df['budget'] = pd.to_numeric(df['budget'], errors='coerce')
df['revenue'] = pd.to_numeric(df['revenue'], errors='coerce')
```

```python
df = df[df['budget'] > 0]
```

```python
df['profit'] = df['revenue'] - df['budget']
df['roi'] = (df['profit'] / df['budget']) * 100

df['roi'].replace([np.inf, -np.inf], np.nan, inplace=True)
df = df.dropna(subset=['roi'])
```

```python
top_movie = df.loc[df['profit'].idxmax()]

print("\n Highest Profit Movie\n")
print("Movie:", top_movie['title'])
print("Profit:", top_movie['profit'])
print("Director:", top_movie['director'])
print("Actors:", top_movie['actors'])
print("Producers:", top_movie['producers'])
```

# This data has information about movies made in different languages. Which language has the highest average ROI (return on investment)?

```python

lang_roi = df.groupby('original_language')['roi'].mean().sort_values(ascending=False)

print("\n Top Language by ROI\n")
print(lang_roi.idxmax())
```

# Find out the unique genres of movies in this dataset.

```python
def extract_genres(x):
    genres = []
    try:
        data = ast.literal_eval(x)
        for i in data:
            genres.append(i.get('name'))
    except:
        pass
    return genres

df['genre_list'] = df['genres'].apply(extract_genres)

unique_genres = set()
for g in df['genre_list']:
    unique_genres.update(g)

print("Unique Genres\n")
print(unique_genres)
```

# Make a table of all the producers and directors of each movie. Find the top 3 producers who have produced movies with the highest average RoI?

```python
df_exp = df.explode('producers')
df_exp = df_exp.dropna(subset=['producers'])

producer_roi = df_exp.groupby('producers')['roi'].mean()

top3_producers = producer_roi.sort_values(ascending=False).head(3)

top3_producers = top3_producers.reset_index()
top3_producers.columns = ['Producer', 'Average ROI']

print("Top 3 Producers by ROI\n")
print(top3_producers)
```

# Which actor has acted in the most number of movies? Deep dive into the movies, genres and profits corresponding to this actor.

```python
actor_counter = Counter()

for actors in df['actors']:
    actor_counter.update(actors)

top_actor = actor_counter.most_common(1)[0][0]

print("Actor with Most Movies:")
print( top_actor)

actor_movies = df[df['actors'].apply(lambda x: top_actor in x)]

print("\nMovies of this actor:\n")
print(actor_movies[['title','profit']])
```

# Top 3 directors prefer which actors the most?

```python
top_directors = df['director'].value_counts().head(3).index

print("Top Directors & Their Favorite Actors\n")

for director in top_directors:
    director_movies = df[df['director'] == director]
    actor_count = Counter()

    for actors in director_movies['actors']:
        actor_count.update(actors)

    print(f"\nDirector: {director}")
    print("Top Actors:", actor_count.most_common(3))
```
