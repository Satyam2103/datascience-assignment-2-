#question 6: which actor has acted in the most number of movies? deep dive into the movies,genres and profits corresponding to this actor.

```python
data.isnull().sum()
```

```python
df6=data.dropna(subset=['genres','cast']).reset_index(drop=True)
```

```python
df6
```

```python
df6=df6[df6['budget']>10000]
```

```python
df6['profit']=pd.DataFrame((df6['revenue']-df6['budget']),columns=['ROI'])
```

```python
df6['cast']=df6['cast'].apply(convert)
```

```python
df6['cast'][0]
```

```python
actor=[]
for i in range(len(df6["cast"])):
  l=[]
  for j in range(len(df6["cast"].iloc[i])):
    l.append(df6["cast"].iloc[i][j]["name"])
  actor.append(l)
```

```python
df6["actor"]=actor
```

```python
df6['genres']=df6['genres'].apply(convert)
```

```python
df6['genres_name']=df6['genres'].apply(lambda x: [i['name'] for i in x] if(isinstance(x,list)) else None)
```

```python
table6=df6[['actor','original_title','profit','genres_name']]
```

```python
table6
```

```python
table6=table6.explode('actor')
```

```python
table6
```

```python
actors_movie=table6.groupby('actor')['original_title'].nunique()
```

```python
actors_movie.sort_values(ascending=False)
```

```python
top_actor=actors_movie.idxmax()
```

```python
table_top_actor=table6[table6['actor']==top_actor].reset_index(drop=True)
```

```python
table_top_actor
```

```python
top_actor_movie=[]
top_actor_genres=set()
profit=0
for i in range(len(table_top_actor)):
  top_actor_movie.append(table_top_actor["original_title"][i])
  profit+=table_top_actor['profit'][i]
  for j in range(len(table_top_actor['genres_name'][i])):
    top_actor_genres.add(table_top_actor['genres_name'][i][j])
```

```python
#top actor
top_actor
```

```python
#top actor's movies
top_actor_movie
```

```python
#top actor's profit
print(profit)
```

```python
#top actor's genres
list(top_actor_genres)
```
